### Ingestion & Parsing of Word Documents

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

#### Using Docx2txtLoader

```python
try:
    doc_loader = Docx2txtLoader(file_path="data/word_docs/Project Report.docx")
    docs = doc_loader.load()
    print(f"No of Documents: {len(docs)}")
    print(f"Preview of content from Document 1 : {docs[0].page_content[0:100]}")
    print(f"Metadata from Document 1: {docs[0].metadata}")

except Exception as e:
    print(f"Error occurred: {e}")    
```

#### Using UnstructuredWordDocumentLoader

```python
try:
    un_loader =UnstructuredWordDocumentLoader(file_path="data/word_docs/Project Report.docx", mode="elements") #mode="single(default)", "multi", "all"
    un_docs = un_loader.load()  #mode="elements" retains same structure of the document
    print(f"No of Documents: {len(un_docs)}")
    print(f"Preview of content from Document 1 : {un_docs[0].page_content[0:100]}")
    print(f"Metadata from Document 1: {un_docs[0].metadata}")

except Exception as e:
    print(f"Error occurred: {e}")   
```
