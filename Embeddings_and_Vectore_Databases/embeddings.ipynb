### Embeddings

* Think of embeddings as a way to translate words into a language that computers understand - numbers.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#let's consider each word is represented as a vector of 2 dimension.

word_embeddings={
    "cat":[0.8, 0.6],
    "kitten":[0.75, 0.65],
    "dog":[0.5, 0.6],
    "puppy":[0.55, 0.67],
    "car":[0.1, 0.3], 
    "truck":[0.12, 0.34]
}
```

```python
plt.subplots(figsize=(8,4))
for word, coord in word_embeddings.items():
    plt.scatter(coord[0], coord[1])
    plt.annotate(word, (coord[0], coord[1]))
plt.title("Embedding Visualization") 
plt.xlabel("dimension-1") 
plt.ylabel("dimension-2")   
```

#### Measuring Cosine Similarity

```python
def cosine_similarity(vec_1, vec_2):
    """
    cosine similarity measures the angle betwen two vectors.
    - Result close to 1: very similar
    - Result close to -1: opposite
    - Result clsoe to 0: not similar
    """
    dot_product = np.dot(vec_1, vec_2)
    norm_1 = np.linalg.norm(vec_1)
    norm_2 = np.linalg.norm(vec_2)
    similarity = dot_product/(norm_1 * norm_2)
    return similarity
```

```python
cosine_similarity([0.8, 0.6], [0.75, 0.65])
```
